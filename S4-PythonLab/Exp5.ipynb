```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(2, 11)
y = x**4 + 5

plt.plot(x, y)
plt.title('y = x**4 + 5')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(2, 11)
y = x**2

plt.subplot(1,2,1)
plt.plot(x,y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title('Normal')
plt.grid(True)

plt.subplot(1,2,2)
plt.stem(x, y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title('Stem')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

continents = ['Africa', 'Asia', 'Europe', 'North America', 
              'Oceania', 'South America', 'Soviet Union']
areas = [11.7, 10.4, 1.9, 9.4, 3.3, 6.9, 7.9]

plt.bar(continents, areas)
plt.xticks(rotation = 45)
plt.xlabel('Continents')
plt.ylabel('sq miles')
plt.title("Bar Graph")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 2, 1000)
input_signal = np.sin(2 * np.pi * t)
output_signal = np.abs(input_signal)

plt.plot(t, input_signal, label='Input')
plt.plot(t, output_signal, label='Output')
plt.title('Full Wave Rectifier')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

heights = [137.5, 142.5, 147.5, 152.5]  # mid-points
frequency = [4, 12, 16, 8]
bins = [135, 140, 145, 150, 155]

plt.hist(np.repeat(heights, frequency), bins=bins, edgecolor='black')
plt.title('Student Height Distribution')
plt.xlabel('Height (cm)')
plt.ylabel('Number of Students')
plt.grid(True)
plt.show()
```
